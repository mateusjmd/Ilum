```python
import plotly.graph_objects as go

wavelength = []
absorbance = []

file_name = ['sintese1.txt', 'sintese2.txt']

for name in file_name:
    with open(f'{name}', 'r') as file:
        W = []
        A = []

        for line in file:
            w, a = line.strip().split(',')
            W.append(float(w))
            A.append(float(a))

        wavelength.append(W)
        absorbance.append(A)

# COMPRIMENTO DE ONDA MÁXIMO
max_w = (
    wavelength[0][absorbance[0].index(max(absorbance[0]))],
    wavelength[1][absorbance[1].index(max(absorbance[1]))]
)
print(f'W_máx_A_1 = {max_w[0]} \nW_máx_A_2 = {max_w[1]}')

# FWMH DA AMOSTRA 1
index2_a1 = None
for idx, val in enumerate(absorbance[0]):
    if val <= 0.25:
        index2_a1 = idx
        break

if index2_a1 is not None:
    FWMH_A1 = wavelength[0][index2_a1] - wavelength[0][0]
else:
    FWMH_A1 = None
print(f'FWMH_A1 = {FWMH_A1}')

# FWMH DA AMOSTRA 2
index1_a2 = None
index2_a2 = None

for idx, val in enumerate(absorbance[1]):
    if index1_a2 is None and val <= 0.25:
        index1_a2 = idx
    elif index1_a2 is not None and val <= 0.25:
        index2_a2 = idx
        break

if index1_a2 is not None and index2_a2 is not None:
    FWMH_A2 = wavelength[1][index2_a2] - wavelength[1][index1_a2]
else:
    FWMH_A2 = None
print(f'FWMH_A2 = {FWMH_A2}')

# PLOT COM PLOTLY
colors = ['blue', 'red']

fig = go.Figure()

for i in range(len(wavelength)):
    fig.add_trace(go.Scatter(
        x=wavelength[i],
        y=absorbance[i],
        mode='lines',
        name=f'Amostra {i+1}',
        line=dict(color=colors[i])
    ))

fig.update_layout(
    title='UV-Vis Absorbance Spectrum of Silver Nanoparticles',
    xaxis_title='Wavelength (nm)',
    yaxis_title='Absorbance (a.u.)',
    template='plotly_white'
)

fig.show()
```
